```python

print('Missing values')
for col in df.columns:
    missing = df[col].isnull().sum()
    if missing > 0:
        print(f"{col}: {missing} missing values")


print('\n Duplicate values')
for col in df.columns:
    duplicates = df[col].duplicated().sum()
    if duplicates > 0:
        print(f"{col}: {duplicates} duplicate values")
```

```python
#columns category with null values

df['Vehicle Type'] = df['Vehicle Type'].fillna("Not Specified")
df['Reason for cancelling by Customer'] = df['Reason for cancelling by Customer'].fillna("Not Applicable")
df['Driver Cancellation Reason'] = df['Driver Cancellation Reason'].fillna("Not Applicable")
df['Incomplete Rides Reason'] = df['Incomplete Rides Reason'].fillna("Not Applicable")
df['Payment Method'] = df['Payment Method'].fillna("Unknown")
df[ 'Cancelled Rides by Customer'] = df[ 'Cancelled Rides by Customer'].fillna("Not Applicable")
df[ 'Cancelled Rides by Driver'] = df[ 'Cancelled Rides by Driver'].fillna("Not Applicable")
df[ 'Incomplete Rides'] = df[ 'Incomplete Rides'].fillna("Not Applicable")

print('\n Missing values after imputation')
print(df.isnull().sum())
```

*clean and numbers of unique values of categorical columns*



```python
# clean and numbers of unique values of categorical columns

df['Avg VTAT'] = df['Avg VTAT'].fillna(0)
df['Avg CTAT'] = df['Avg CTAT'].fillna(0)
df['Booking Value'] = df['Booking Value'].median()
df['Driver Ratings'] = df['Driver Ratings'].median()
df['Customer Rating'] = df['Customer Rating'].median()
df['Ride Distance'] = df['Ride Distance'].median()

print('\n Missing values after imputation')
print(df.isnull().sum())
```



Driver Cancellation Reasons Analysis

*Analyzing the cancellation reasons in our dataset provides valuable insights into customer and driver behavior. By examining why trips were canceled, we can identify the most common causes and prioritize areas for improvement. The chart below presents the distribution of driver cancellation reasons in percentages, highlighting which factors contribute most frequently to cancellations. Understanding these patterns allows us to develop strategies to reduce cancellations and enhance the overall customer experience.*



```python
# filter real values (without "Not Applicable")

driver_reasons = df['Driver Cancellation Reason']
driver_reasons = driver_reasons[driver_reasons != "Not Applicable"]

# calculate percentages
driver_counts = driver_reasons.value_counts(normalize=True) * 100
driver_counts = driver_counts.sort_values(ascending=False)  

# create DataFrame for hue
driver_df = driver_counts.reset_index()
driver_df.columns = ['Reason', 'Percentage']

# plot horizontal bar chart in percentage with hue
plt.figure(figsize=(12,6))
sns.barplot(
    x='Percentage',
    y='Reason',
    data=driver_df,
    hue='Reason',
    dodge=False,
    palette='viridis',
    legend=False
)

plt.title("Driver Cancellation Reasons (%)")
plt.xlabel("Percentage of Cancellations (%)")
plt.ylabel("Cancellation Reason")

# add percentage labels
for i, v in enumerate(driver_df['Percentage']):
    plt.text(v + 0.5, i, f"{v:.1f}%", va='center')

plt.show()
```

*the first top 5 reasons for driver cancellations are:*

- Customer related issue
- The customer was coughing/ sick
- Personal and Car related issues
- More than permitted people in the vehicle

*I noticed that there is an inconsistency in the values, for example, the customer cancels the trip because the driver does not go to the address, representing 22.5%, perhaps the drivers' reasons for canceling are due to problems related to the customer, isn't that confusing?*

```python
print(df.columns)

# --- Drivers ---
driver_reasons = df['Driver Cancellation Reason']
driver_reasons = driver_reasons[driver_reasons != "Not Applicable"]
driver_counts = driver_reasons.value_counts(normalize=True) * 100
driver_df = driver_counts.reset_index()
driver_df.columns = ['Reason', 'Driver_Percentage']

# --- Customers ---
customer_reasons = df['Reason for cancelling by Customer']
customer_reasons = customer_reasons[customer_reasons != "Not Applicable"]
customer_counts = customer_reasons.value_counts(normalize=True) * 100
customer_df = customer_counts.reset_index()
customer_df.columns = ['Reason', 'Customer_Percentage']

# --- Merge for comparison ---
comparison_df = pd.merge(driver_df, customer_df, on='Reason', how='outer').fillna(0)
comparison_df['Gap'] = np.abs(comparison_df['Customer_Percentage'] - comparison_df['Driver_Percentage'])
comparison_df = comparison_df.sort_values('Customer_Percentage', ascending=False)
y_positions = range(len(comparison_df))

# --- Define colors based on gap ---
# higher gap -> darker color
max_gap = comparison_df['Gap'].max()
colors = ['#ff9999' if gap > max_gap*0.5 else '#ffcccc' for gap in comparison_df['Gap']]
colors_customer = ['#66b3ff' if gap > max_gap*0.5 else '#99ccff' for gap in comparison_df['Gap']]

plt.figure(figsize=(12,8))

# customer bars
plt.barh(
    [y + 0.2 for y in y_positions], 
    comparison_df['Customer_Percentage'], 
    height=0.4, 
    color=colors_customer, 
    label='Customer'
)

# driver bars
plt.barh(
    [y - 0.2 for y in y_positions], 
    comparison_df['Driver_Percentage'], 
    height=0.4, 
    color=colors, 
    label='Driver'
)

# add percentage labels
for i, (c, d) in enumerate(zip(comparison_df['Customer_Percentage'], comparison_df['Driver_Percentage'])):
    plt.text(c + 0.5, i + 0.2, f"{c:.1f}%", va='center', fontsize=9)
    plt.text(d + 0.5, i - 0.2, f"{d:.1f}%", va='center', fontsize=9)

plt.yticks(y_positions, comparison_df['Reason'])
plt.xlabel("Percentage of Cancellations (%)")
plt.ylabel("Cancellation Reason")
plt.title("Comparative Cancellation Reasons: Customer vs Driver (Gaps Highlighted)")
plt.legend()
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
